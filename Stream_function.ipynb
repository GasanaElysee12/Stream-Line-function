# Here we have our velocities $u=x^2,\ v=-2xy-1,\ w=0$. we are going to solve the streamfunction $\psi(x,y)$ and plot strem lines

* $ u=\frac{\partial \psi(x,y}{\partial y},\  v=-\frac{\partial \psi(x,y}{\partial x}$ 

```python
import matplotlib.pyplot as plt
import numpy as np


#define the function Stream function in terms of stream function and x
def Stream(psi,x,c=0.0):
    
    y=(psi-x-c)/x**2
    return y

#define x coordinate
Y_value=[]

X=np.linspace(1,5,8)

#Initialize c=0.0
c=0.0

#based on how psi is constant along the stream line, let compute y for psi in range(6)
for psi in range(6):
    y_per_psi=[]
    for x in X:
        y_per_psi.append(Stream(psi,x,c=0.0))
    Y_value.append(y_per_psi)
    
```

# Let plot $y=\frac{\psi (x,y)-x-c}{x^2}$ for $c=0.0$

```python
for i in range(6):
    plt.plot(X,Y_value[i],label=r'$\psi(x,y)=%s$'%i)
    plt.xlabel('X-value')
    plt.ylabel('Y-value')
    plt.grid('on')
    plt.title('Stream lines')
    
    plt.legend(loc='upper right', fancybox=True, bbox_to_anchor=(1.6,0.7),shadow=True,facecolor='k',labelcolor='g',fontsize='xx-large')
    ind=0
    #Let's draw the arrows that show the stream functions
    while ind<len(X)-1:
        plt.quiver(X[ind],0.1+Y_value[i][ind], X[ind+1]-X[ind],Y_value[i][ind+1]-Y_value[i][ind])
        ind+=1
plt.show()
```
